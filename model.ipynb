```python

import torch 
import cv2
import os
from matplotlib import pyplot as plt
from torch.utils.data import DataLoader, TensorDataset
import numpy as np
from torch import optim
import torch.nn.functional as F
```

```python
#this function resizes an image
def resizer(image):
    resized = cv2.resize(image, (400,400))
    return resized

#this function extracts all the images from a folder
def load_data(data, label):
    images = []
    Y = []
    files = os.listdir(data)
    print(data)
    for file in files:
        image = cv2.imread(data+"/"+file, cv2.IMREAD_UNCHANGED)
        image = resizer(image)
        images.append(image)
        Y.append(label)
    return images, Y
  
# print('boo')
# train_x, train_y = load_data(data_train)
# print('done')
folders = os.listdir('./data')
for folder in folders:
    print(folder)
    images, y = load_data('./data/'+folder, folder)
    plt.imshow(images[5])
    plt.show()
    print(y[5])
    break
    
# plt.imshow(train_x[0])
# plt.show()
```
